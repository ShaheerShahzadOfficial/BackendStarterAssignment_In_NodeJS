```python
import re
import nltk
from nltk.corpus import stopwords
import emoji

# Download NLTK stopwords if not already downloaded
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')

# Get the original text
cleaned_text = text

# 1. Convert to lowercase
cleaned_text = cleaned_text.lower()
print("--- After lowercasing ---")
print(cleaned_text[:500]) # Display first 500 characters

# 2. Remove HTML tags
cleaned_text = re.sub(r'<.*?>', '', cleaned_text)
print("\n--- After removing HTML tags ---")
print(cleaned_text[:500])

# 3. Remove emojis
cleaned_text = emoji.replace_emoji(cleaned_text, replace='')
print("\n--- After removing emojis ---")
print(cleaned_text[:500])

# 4. Remove punctuation
cleaned_text = re.sub(r'[^\w\s]', '', cleaned_text)
print("\n--- After removing punctuation ---")
print(cleaned_text[:500])

# 5. Remove numbers
cleaned_text = re.sub(r'\d+', '', cleaned_text)
print("\n--- After removing numbers ---")
print(cleaned_text[:500])

# 6. Remove stopwords
stop_words = set(stopwords.words('english'))
words = cleaned_text.split()
filtered_words = [word for word in words if word not in stop_words]
cleaned_text = ' '.join(filtered_words)
print("\n--- After removing stopwords ---")
print(cleaned_text[:500])

# 7. Remove extra spaces
cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()
print("\n--- After removing extra spaces ---")
print(cleaned_text[:500])

print("\n\n--- Final Cleaned Text ---")
print(cleaned_text)
```
